```python
epoch_type = 'evoked'
# epoch_type = 'pseudo-rs'

all_participants = ['VP07','VP08','VP09', 'VP10','VP11','VP12','VP13','VP14','VP18','VP19','VP20','VP22','VP23','VP24','VP25','VP26','VP27','VP28','VP29','VP30','VP31','VP32','VP33','VP35','VP36','VP37']
participant = all_participants[1]

path = '/media/nicolas.bruno/63f8a366-34b7-4896-a7ce-b5fb4ee78535/Nico/MW_eeg_data/minmarker/' #icm-linux
# path = '/Users/nicobruno/ownCloud/MW_eeg_data/minmarker/' #nico-mac
```

```python
rejection = []

for i in all_participants:
    participant = i
    
    print('')
    print('#########################################')
    print('Preprocessing participant {}'.format(participant))
    print('#########################################')
    print('')
    

    folder = path + participant +'/'
    raw_filtered = mne.io.read_raw(folder + participant + 'filt_raw.fif', preload = True)

    events, event_dict = utils.make_correct_labels(raw_filtered, nb_prev_trials = 7)

    # Create Epochs <a name="epochs"></a>

    epochs = utils.create_epochs(epoch_type = epoch_type, raw = raw_filtered, events =  events, event_id = event_dict)
    epochs.save(folder +  participant + '_' + epoch_type + '_' +  '_epo.fif', overwrite = True)

    # Reject Bad Epochs and Channels <a name="reject"></a>

    ar = AutoReject(thresh_method = 'random_search',n_interpolate = np.array([1, 4, 8, 16]) ,random_state=42)
    epochs_clean = ar.fit_transform(epochs)  
    reject = get_rejection_threshold(epochs)  

    epochs_clean.save(folder +  participant + '_' + epoch_type + '_' +  'ar_clean_epo.fif', overwrite = True)
    
    rejection.append((len(epochs) - len(epochs_clean)) *100/len(epochs))
    
print(rejection)
```

```python
probe = ['PC', 'SC']
mind = ['on-task', 'about-task','distracted','dMW', 'sMW']
go = ['go', 'nogo']
correct = ['correct', 'incorrect']

for i in all_participants:
    participant = i
    
    print('')
    print('#########################################')
    print('Preprocessing participant {}'.format(participant))
    print('#########################################')
    print('')
    

    folder = path + participant +'/'
    epochs_rereferenced = mne.read_epochs(folder +  participant + '_' + epoch_type + '_' +  'ar_rereferenced_epo.fif')    
    

    list_epochs_subtracted = []
    for i in probe:
        for j in mind:
            for k in go:
                for l in correct:
                    try:
                        condition = [i + '/' + j + '/' + k + '/' + l] 
                        epochs_condition = epochs_rereferenced[condition].copy()
                        if len(epochs_condition) > 0:
                            list_epochs_subtracted.append(epochs_condition.subtract_evoked())
                    except:
                        pass


    epochs_subtracted = mne.concatenate_epochs(list_epochs_subtracted)
    epochs_subtracted.save(folder +  participant + '_' + epoch_type + '_' +  'ar_subtracted_epo.fif', overwrite = True)
```
